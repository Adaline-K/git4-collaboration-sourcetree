```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
df1= pd.DataFrame(np.random.randn(100,3),
                  index=pd.date_range('1/1/2018', periods=100),
                  columns=['A','B','C']).cumsum()
df1.tail()
```

```

df1.plot()
plt.title("Pandas plot method example")
plt.xlabel("Time")
plt.ylabel("Data")
plt.show()
```

```
#plot method kind(bar, pie, hist, kde, box, scatter, area)
iris = sns.load_dataset("iris")    # iris dataset
titanic = sns.load_dataset("titanic")    # tatanic dataset

iris.sepal_length[:15].plot(kind='bar', rot=0)
plt.title("flower")
plt.xlabel("Data")
plt.ylabel("length")
plt.show()
```

```
iris[:5].plot.bar(rot=0)
plt.title("Iris data Bar Plot")
plt.xlabel("Data")
plt.ylabel("Each Feature Value")
plt.ylim(0, 7)
plt.show()
```

```
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"
df2
```

```
df2.plot.bar(rot=0)
plt.title("Each Feature average")
plt.xlabel("Average")
plt.ylabel("Species")
plt.ylim(0,8)
plt.show()
```

```
df2.T.plot.bar(rot=0)
plt.title("Each Feature average")
plt.xlabel("Feature")
plt.ylabel("Average")
plt.show()
```

```
def sinplot(flip=1):
  x= np.linspace(0, 14, 100)
  for i in range(1,7):
    plt.plot(x, np.sin(x +i *.5) * (7 -1) * flip)
```
